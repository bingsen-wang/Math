```
from numpy import linspace,cos,arccos,pi
from matplotlib import animation,pyplot,cm
pyplot.rcParams.update({'font.family':"sans serif",'animation.html':'jshtml',
                     'mathtext.fontset':'cm'})
def Ln(x,Ln1,Ln2):
  return x*Ln1+Ln2

Nf=40
clst = list(cm.tab20.colors) #colors of lines
x = linspace(-1,1,1000)
Ln_lst = [x*0+2,x]
[Ln_lst.append(Ln(x,Ln_lst[i-1],Ln_lst[i-2])) for i in range(2,Nf+1)]
fig = pyplot.figure(figsize=([9,9]))
ax = fig.add_subplot(111,xlim=(-1,1),ylim=(-10,10))
lines=[ax.plot([],[],lw=1.5,c=clst[i%len(clst)])[0] for i in range(len(Ln_lst))]
def animate(ii):
  # ii=2*Nf-1
  i=ii%Nf
  k=ii//Nf
  if k==0:
    ax.set_title(f"Lucas Polynomial of Order ${i}$",size=18)
    lines[i-1].set_data([],[])
    lines[i].set_data(x,Ln_lst[i])
  else:
    if i==0:
      ax.set_title(f"Lucas Polynomials of Orders $0-{Nf}$",size=18)
    [lines[j].set_data(x[:(i+1)*len(x)//Nf],Ln_lst[j][:(i+1)*len(x)//Nf]) for j in range(Nf)]
  return
anim = animation.FuncAnimation(fig, animate, frames=Nf*2, interval=50)
anim.save("LucasPolynomials.mp4", writer=animation.FFMpegWriter(fps=5), dpi = 300)
```
