```
from matplotlib import pyplot,animation,cm,patches
import numpy as np
pyplot.rcParams.update({"font.family":"sans-serif", "mathtext.fontset":"cm"})
nc=10;r0=1
rs = [r0*(np.sqrt(2)/2)**k for k in range(nc)]
cl = list(cm.tab10.colors) #colors of lines
fig = pyplot.figure(figsize=([9,9]))
ax = fig.add_axes([0,0,1,1],xlim=[-1.1,1.1],ylim=[-1.1,1.1])
ax.axis('off')
def animate(ii):
  k=ii%9;i=ii//9
  if k==0:
    ax.add_patch(patches.Circle((0,0),rs[i],color=cl[ii%len(cl)],fill=True))
  else:
    ax.add_patch(patches.Rectangle((-rs[i]/np.sqrt(2),-rs[i]/np.sqrt(2)),rs[i]*np.sqrt(2),rs[i]*np.sqrt(2),
                                   angle=(k-1)*90/8,rotation_point=(0,0),color=cl[ii%len(cl)],fill=True))
  return
ax.text(np.average(ax.get_xlim()),ax.get_ylim()[0]*.99+ax.get_ylim()[1]*.01,r'@eeanimation',size=12,c='.2',alpha=.3,ha='center',va='bottom')
anim = animation.FuncAnimation(fig, animate, frames=9*nc, interval=50)
anim.save("squaresCicles.mp4",writer=animation.FFMpegWriter(fps=9),dpi=200)
```
