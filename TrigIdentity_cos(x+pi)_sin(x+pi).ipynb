```
import matplotlib.pyplot as plt
plt.rcParams.update({'font.family':"sans serif",'mathtext.fontset':'cm'})
fig = plt.figure(figsize=([9,9]))
ax= fig.add_axes([0,0,1,1])
ax.axis('off')
ax.text(.5,.9,'Trig Identities: $\cos(x\pm\pi)$ and $\sin(x\pm\pi)$',size=30,ha='center',va='center')
ax.text(.5,.8,'$\cos(x+y) = \cos x\cos y - \sin x\sin y$\n $\sin(x+y) =\sin x\cos y +\cos x\sin y$',
        size=30,ha='center',va='top',multialignment='right',
         bbox=dict(boxstyle='ellipse',pad=0,facecolor='w',edgecolor='tab:orange',lw=5))
ax.text(.5,.58,'$y=\pm\pi$',c='tab:blue',size=25,bbox=dict(boxstyle='larrow',pad=.1,facecolor='w',edgecolor='tab:orange',lw=3),
        ha='center',va='center',rotation=90)
ax.text(.5,.45,'$\cos(x\pm\pi) = \cos x\cos(\pm\pi) - \sin x\sin(\pm\pi)$ \n\n$\sin(x\pm\pi) =\sin x\cos(\pm\pi) +\cos x\sin(\pm\pi)$',
        size=25,ha='center',va='top',multialignment='right',
        bbox=dict(boxstyle='round',pad=.8,facecolor='w',edgecolor='tab:blue',lw=3))
ax.annotate("$=0$", xy=(.72, .4), xytext=(.82, .46), size=12, c='tab:red',
            arrowprops=dict(arrowstyle="<-", color="tab:red", lw=2))
ax.annotate("$=0$", xy=(.72, .3), xytext=(.82, .36), size=12, c='tab:red',
            arrowprops=dict(arrowstyle="<-", color="tab:red", lw=2))
ax.annotate("", xy=(.45, .41), xytext=(.59, .41), arrowprops=dict(arrowstyle="|-|", color="k", lw=2))
ax.text(.52,.41,'$=-1$',size=18,va='top',ha='center')
ax.annotate("", xy=(.45, .31), xytext=(.59, .31), arrowprops=dict(arrowstyle="|-|", color="k", lw=2))
ax.text(.52,.31,'$=-1$',size=18,va='top',ha='center')

ax.text(.5,.2,'$\cos(x\pm\pi) = -\cos x$\n$\sin(x\pm\pi) =-\sin x$',
        size=35,ha='center',va='top',multialignment='left',
         bbox=dict(boxstyle='roundtooth',pad=.5,facecolor='w',edgecolor='tab:green',lw=6))
fig.savefig('trigIdentities_plus_minus_pi.png',dpi=300)
```
