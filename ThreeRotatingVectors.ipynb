#Python Code

```
from matplotlib import pyplot,animation,cm
import numpy as np
pyplot.rcParams.update({"font.family":"sans-serif", "mathtext.fontset":"cm"})
Nf,r,= 1500,.5; theta = np.linspace(0,2*np.pi,Nf)
f = [r*np.exp(5j*theta),.4*r*np.exp(17j*theta),.25*r*np.exp(29j*theta)]
fsum = [sum(f[:k+1]) for k in range(len(f))]
clst = list(cm.tab10.colors) #colors of lines
fig = pyplot.figure(figsize=([9,9]))
ax = fig.add_axes([0,0,1,1],xlim=[-1,1],ylim=[-1,1]); ax.axis('off')
lines=[ax.plot([],[],'-',lw=2, c=clst[k])[0] for k in range(3)]
def animate(i):
  ax.plot(fsum[-1][i-1:i+1].real,fsum[-1][i-1:i+1].imag,lw=2.5, c=clst[(i//5)%10])
  lines[0].set_data([0,f[0][i].real],[0,f[0][i].imag])
  [lines[k+1].set_data([fsum[k][i].real,fsum[k+1][i].real],[fsum[k][i].imag,fsum[k+1][i].imag]) for k in range(2)]
  return
anim = animation.FuncAnimation(fig, animate, frames=Nf, interval=50)
anim.save("threevectors.mp4",writer=animation.FFMpegWriter(fps=60),dpi=200)
```
