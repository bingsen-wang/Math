```
from numpy import linspace,cos,sin,arccos,pi,abs,exp,sum
from math import factorial
from matplotlib import animation,pyplot,cm
pyplot.rcParams.update({'font.family':"sans serif",'animation.html':'jshtml',
                     'mathtext.fontset':'cm'})
def infs_exp(x,n):
  return (1+x/n)**n

def str_yn(k):
  if k==1:
    yn='1+x'
  else:
    yn = r'\left(1+\frac{x}{'+str(k)+r'}\right)^{'+str(k)+'}'
    # yn = str(k)
  return yn

Nf=60
clst = list(cm.tab10.colors) #colors of lines
xn,xm=-5,5
x = linspace(xn,xm,500)
Npf = len(x)//10
tsn_lst=[(infs_exp(x,i)) for i in range(1,Nf+1)]
fig = pyplot.figure(figsize=([12,12]))
ax = fig.add_subplot(111,xlim=(xn,xm),ylim=(-10,100))
ax.set_title(r'$\left(1+\frac{x}{n}\right)^n$ vs $e^x$',size=45)
ax.set_xlabel('$x$',size=28),ax.set_ylabel('$y$',size=28)
ax.plot(x,exp(x),'--k',lw=2)
txt = ax.text(0,80,'',size=45,ha='center',va='center')
lines=ax.plot([],[],lw=5)[0]
def animate(ii):
  # ii=60*Npf-1
  i=ii%Npf
  k=ii//Npf+1
  if i==0:
    txt.set_text('$y='+str_yn(k)+'$')
    txt.set_color(clst[(k-1)%len(clst)])
  lines.set_data(x[:(i+1)*10],tsn_lst[k-1][:(i+1)*10])
  lines.set_color(clst[(k-1)%len(clst)])
  return
anim = animation.FuncAnimation(fig, animate, frames=Nf*Npf, interval=50)
anim.save("InfiniteSeries_expx.mp4",writer=animation.FFMpegWriter(fps=50),dpi=200)
```
