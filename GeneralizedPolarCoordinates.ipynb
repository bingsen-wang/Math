```
import numpy as np
from matplotlib import animation,pyplot,cm
pyplot.rcParams.update({'font.family':"sans serif",'animation.html':'jshtml',
                     'mathtext.fontset':'cm'})
Nf=250
theta = np.linspace(0,2*np.pi,1000)
clst = list(cm.tab10.colors)#colors of lines
fig = pyplot.figure(figsize=([9,9]),facecolor='none')
ax0 = fig.add_axes((.05,0,.9,.9),projection='polar',fc='none')
# ax0.axis('off')
ax0.set_rmax(1.1)
txt = ax0.text(0,1.1,'',ha='center',va='center',size=25)
rm =1
rn = list(np.linspace(1,0,Nf))+list(np.linspace(0,1,Nf))
lines=[ax0.plot([],[],ls=['-','--'][i],color=clst[i],lw=[3,1.5][i])[0] for i in range(2)]
texts=[[ax0.text(0,0,'',c=clst[k],ha='center',size=10) for i in range(Nf)]for k in range(4)]
def animate(i):
  # i=50
  r = ((rm+rn[i])*.5 + .5*(rm-rn[i])*np.cos(6*theta))
  ax0.set_title(rf'$\rho(\theta)={.5*(rm+rn[i]):.2f} + {.5*(rm-rn[i]):.2f}\cos(6\theta)$',size=25)
  r0 = (rm+rn[i])*.5+theta*0
  lines[0].set_data(theta,r)
  lines[1].set_data(theta,r0)
  return
anim = animation.FuncAnimation(fig, animate, frames=2*Nf, interval=50)
anim.save("SinusoidalInPolar.mp4", writer=animation.FFMpegWriter(fps=50), dpi=200)
```
