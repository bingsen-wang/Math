```
from numpy import linspace,sin,arccos,pi
from matplotlib import animation,pyplot,cm
pyplot.rcParams.update({'font.family':"sans serif",'animation.html':'jshtml',
                     'mathtext.fontset':'cm'})
def W(x,n):
  return sin((n+.5)*arccos(x))/sin(arccos(x)*.5)

Nf=40
clst = list(cm.tab20.colors) #colors of lines
x = linspace(-1,1,1000)
Wn_lst = [W(x,k+1) for k in range(Nf)]
fig = pyplot.figure(figsize=([9,9]))
ax = fig.add_subplot(111,xlim=(-1,1),ylim=(-3,3))
lines=[ax.plot([],[],lw=1.5,c=clst[i%len(clst)])[0] for i in range(len(Wn_lst))]
def animate(ii):
  # ii=2*Nf-1
  i=ii%Nf
  k=ii//Nf
  if k==0:
    ax.set_title(f"Chebyshev Polynomial of the Second Kind: Order ${i+1}$",size=18)
    lines[i-1].set_data([],[])
    lines[i].set_data(x,Wn_lst[i])
  else:
    if i==0:
      ax.set_title(f"Chebyshev Polynomials of the Second Kind: Orders $1-{Nf}$",size=18)
    [lines[j].set_data(x[:(i+1)*len(x)//Nf],Wn_lst[j][:(i+1)*len(x)//Nf]) for j in range(Nf)]
  return
anim = animation.FuncAnimation(fig, animate, frames=Nf*2, interval=50)
anim.save("chebyshevPolynomials_FourthKind.mp4", writer=animation.FFMpegWriter(fps=5), dpi = 300)
```
