```
import matplotlib.pyplot as plt
plt.rcParams.update({'font.family':"sans serif",'mathtext.fontset':'cm'})
fig = plt.figure(figsize=([9,9]))
ax= fig.add_axes([0,0,1,1])
ax.axis('off')
ax.text(.5,.9,'Trig Identity: $\cos^2x+\sin^2x$',size=30,ha='center',va='center')
ax.text(.5,.75,'$\cos(x+y) = \cos x\cos y - \sin x\sin y$',
        size=30,ha='center',va='top',multialignment='right',
         bbox=dict(boxstyle='ellipse',pad=.2,facecolor='w',edgecolor='tab:orange',lw=5))
ax.text(.5,.55,'$y=-x$',c='tab:blue',size=35,bbox=dict(boxstyle='larrow',pad=.1,facecolor='w',edgecolor='tab:orange',lw=3),
        ha='center',va='center',rotation=90)
ax.text(.5,.36,'$\cos(x-x) = \cos x\cos (-x) - \sin x\sin (-x)$',
        size=25,ha='center',va='top',multialignment='right',
        bbox=dict(boxstyle='round',pad=1,facecolor='w',edgecolor='tab:blue',lw=3))
ax.annotate("",xy=(.16,.31),xytext=(.34,.31),arrowprops=dict(arrowstyle="|-|", color="k", lw=1.5))
ax.annotate("",xy=(.45,.31),xytext=(.6,.31),arrowprops=dict(arrowstyle="|-|", color="k", lw=1.5))
ax.annotate("",xy=(.7,.31),xytext=(.85,.31),arrowprops=dict(arrowstyle="|-|", color="k", lw=1.5))
ax.text(.25,.31,'$\cos0=1$',c='tab:red',size=14,ha='center',va='top')
ax.text(.525,.31,'$=\cos x$',c='tab:red',size=14,ha='center',va='top')
ax.text(.775,.31,'$=-\sin x$',c='tab:red',size=14,ha='center',va='top')
ax.text(.5,.2,'$\cos^2x +\sin^2x= 1$', size=35,ha='center',va='top',multialignment='left',
         bbox=dict(boxstyle='roundtooth',pad=.5,facecolor='w',edgecolor='tab:green',lw=6))
fig.savefig('trigIdentities_sin^2_plus_cos^2.png',dpi=300)
```
