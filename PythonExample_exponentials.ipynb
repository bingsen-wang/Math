#Python Code
The Python code illustrates how to use ArtistAnimation.

```
from numpy import linspace,logspace
from matplotlib import animation,pyplot,cm,transforms
pyplot.rcParams.update({'font.family':"sans serif",'animation.html':'jshtml',
                     'mathtext.fontset':'cm'})
clst = list(cm.tab20.colors) #colors of lines
x,n = linspace(0,1,500),logspace(0,2,100)
fig,ax = pyplot.subplots(figsize=([9,9]))
ax.set_title('Rotated Lines with Various Exponents',size=28)
ax.set_xlim(-.3,1.3),ax.set_ylim(-.3,1.3)
t1 = ax.text(-.2,1.2,'',size=35,ha='left')
t2 = ax.text(1.2,-.2,'',size=35,ha='right')
def animate(i):
  # i=1
  rot = transforms.Affine2D().rotate_deg_around(.5,.5,i*180/len(n))+ax.transData
  t1.set_text('$y=x^{'+f'{n[i]:.2f}'+'}$')
  t1.set_color(clst[i%len(clst)])
  t2.set_text(r'$y=x^\frac{1}{'+f'{n[i]:.2f}'+'}$')
  t2.set_color(clst[i%len(clst)])
  ax.plot(x,x**n[i],lw=5-4.5*i/len(n),c=clst[i%len(clst)],transform=rot)
  ax.plot(x,x**(1/n[i]),lw=5-4.5*i/len(n),c=clst[i%len(clst)],transform=rot)
  return
anim = animation.FuncAnimation(fig, animate, frames=len(n), interval=50)
anim.save("lines_rotated.mp4", writer=animation.FFMpegWriter(fps=10), dpi = 200)
```
