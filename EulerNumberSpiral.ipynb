#Python Code
The Python code prints Euler's with increaisng number of decimals.

```
from os import linesep
import textwrap
from mpmath import *
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import animation,rc
rc('animation', html='jshtml')
plt.rcParams.update({"font.family" : "sans serif","mathtext.fontset" : "cm"})

#parameters
clst=['lightskyblue','violet','aqua','cornflowerblue','yellow','w',(.1,.1,.1)] #colors dark bgd

Nf = 21*16 +3
f=11.5
mp.dps=Nf-3
estring = 'e='+str(+e)
t=np.linspace(0,1,Nf)
epjt = .12*np.exp(-1j*2*np.pi*f*t+2.1*t)

#set up figures
fig = plt.figure(figsize=(9,9),facecolor=clst[-1])
fig.tight_layout()
ax=fig.add_axes([0, 0, 1, 1],xlim=[-1,1], ylim=[-1,1],fc='none') #no fc
ax.axis('off') #turn off axis frames
txt = ax.text(0,.5,'',size = 50,color=clst[0],va='center')
traj=ax.plot([],[],lw=1,color='lightgray',alpha=0.3)[0]

# animation function. This is called sequentially
def animate(i):
  ax.text(epjt[i].real,epjt[i].imag,estring[i],color=clst[i%6],
          size=10+20*i/Nf,rotation=np.angle(epjt[i],deg=True)-90)
  traj.set_data(epjt[:i+1].real,epjt[:i+1].imag)
  return
anim = animation.FuncAnimation(fig, animate, frames=Nf, interval=200)
# anim #uncomment to generate animation in the output area
# to save the animation, uncomment the following three lines
fn = "EulerNumberSpiral.mp4"
writervideo = animation.FFMpegWriter(fps=25)
anim.save(fn, writer=writervideo,dpi = 120)
```
