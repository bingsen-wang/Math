#Python Code

```
import numpy as np
from matplotlib import pyplot,animation
Nf= 400
theta,r = np.linspace(0,np.pi*2,Nf),np.linspace(3.5,0,Nf)
f = np.sin(theta)**3 + 1j/16*(13*np.cos(theta)-5*np.cos(2*theta)-2*np.cos(3*theta)-np.cos(4*theta))
fig = pyplot.figure(figsize=([9,9]))
ax = fig.add_axes([0,0,1,1],xlim=[-4,4],ylim=[-4,4])
ax.axis('off')
lines=[ax.plot([],[],lw=3,color='tab:red')[0] for _ in range(Nf)]
def animate(i):
  k,ii=i%Nf,i//Nf
  lines[0].set_data(r[0]*f[:k].real,r[0]*f[:k].imag) if ii==0 else lines[k].set_data(r[k]*f.real,r[k]*f.imag)
  return
anim = animation.FuncAnimation(fig, animate, frames=Nf*2, interval=50)
anim.save("Graphs_heart.mp4",writer=animation.FFMpegWriter(fps=50),dpi=200)
```
