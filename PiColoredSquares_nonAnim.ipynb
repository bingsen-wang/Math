```
from matplotlib import pyplot,cm,patches
import numpy as np, mpmath as mp
pyplot.rcParams.update({"font.family":"sans-serif", "mathtext.fontset":"cm"})
nr,nc=100,100; Nf= nr*nc #number of rows, columns, and frames
xylim=[-1,1]
w,h=(xylim[1]-xylim[0])/nc,(xylim[1]-xylim[0])/nr
mp.mp.dps=nr*nc+1; txt=str(mp.pi)
print(len(txt))
cl = list(cm.tab10.colors) #colors of lines
fig = pyplot.figure(figsize=([9,9]))
ax = fig.add_axes([.05,.05,.9,.9],xlim=xylim,ylim=xylim); ax.axis('off')
for row in range(nr):
  for col in range(nc):
    ax.text(xylim[0]+(col+.5)*w,xylim[1]-(row+.5)*h,f'${txt[row*nc+col]}$',size=6,ha='center',va='center')
    if row==0 and col==1:
      ax.add_patch(patches.Rectangle((xylim[0]+col*w,xylim[1]-(row+1)*h),w,h,fc='.5',alpha=.5))
    else:
      ax.add_patch(patches.Rectangle((xylim[0]+col*w,xylim[1]-(row+1)*h),w,h,fc=cl[int(txt[row*nc+col])],lw=0,alpha=.8))
    # elif txt[row*nc+col]=='0': #uncomment to color only circuit digits
    #   ax.add_patch(patches.Rectangle((xylim[0]+col*w,xylim[1]-(row+1)*h),w,h,fc=cl[int(txt[row*nc+col])],lw=0,alpha=.8))
    # else:
    #   continue
ax.text(np.average(ax.get_xlim()),ax.get_ylim()[0]*.99+ax.get_ylim()[1]*.01-.1,r'@eeanimation',size=12,c='.2',alpha=.3,ha='center',va='bottom')
# fig.savefig('pi.png',dpi=600)
```
