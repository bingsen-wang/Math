```
import matplotlib.pyplot as plt
plt.rcParams.update({'font.family':"sans serif",'mathtext.fontset':'cm'})
fig = plt.figure(figsize=([9,9]))
ax= fig.add_axes([0,0,1,1])
ax.axis('off')
ax.text(.5,.9,'Double-Angle Identities: $\cos(2x)$ and $\sin(2x)$',size=32,ha='center',va='center')
ax.text(.5,.75,'$\cos(x+y) = \cos x\cos y - \sin x\sin y$\n $\sin(x+y) =\sin x\cos y +\cos x\sin y$',
        size=30,ha='center',va='top',multialignment='right',
         bbox=dict(boxstyle='ellipse',pad=.1,facecolor='w',edgecolor='tab:orange',lw=5))
ax.text(.5,.52,'$y=x$',c='tab:blue',size=35,bbox=dict(boxstyle='larrow',pad=.1,facecolor='w',edgecolor='tab:orange',lw=3),
        ha='center',va='center',rotation=90)
ax.text(.5,.4,'$\cos(x+x) = \cos x\cos x - \sin x\sin x$\n $\sin(x+x) =\sin x\cos x +\cos x\sin x$',
        size=25,ha='center',va='top',multialignment='right',
        bbox=dict(boxstyle='round',pad=.5,facecolor='w',edgecolor='tab:blue',lw=3))
ax.text(.5,.2,'$\cos(2x) = \cos^2 x - \sin^2 x$\n$\sin(2x) =2\sin x\cos x $',
        size=35,ha='center',va='top',multialignment='left',
         bbox=dict(boxstyle='roundtooth',pad=.5,facecolor='w',edgecolor='tab:green',lw=6))
fig.savefig('trigIdentities_doubleAngle.png',dpi=300)
```
