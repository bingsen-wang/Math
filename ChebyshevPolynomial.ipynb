#Python Code


```
from numpy import linspace
from matplotlib import animation,pyplot,cm

def cheb(n,x):
  if n==0:
    return 1
  elif n==1:
    return x
  else:
    return 2*x*cheb(n-1,x)-cheb(n-2,x)
clst = list(cm.tab20.colors) #colors of lines
x,n = linspace(0,1,500),linspace(1,20,20)
Tn = [cheb(k+1,x) for k in range(len(n))]
fig,ax = pyplot.subplots(figsize=([9,9]))
ax.set_xlim(0,1)
ax.set_ylim(-1,1)
lines = [ax.plot([],[],lw=3,c=clst[i%len(clst)])[0] for i in range(len(n))]
def animate(i):
  lines[i].set_data(x,Tn[i])
  return
anim = animation.FuncAnimation(fig, animate, frames=len(n), interval=50)
anim.save("chebyshevPolynomials.mp4", writer=animation.FFMpegWriter(fps=1), dpi = 200)
```
